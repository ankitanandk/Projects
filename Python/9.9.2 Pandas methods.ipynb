```python
'''4. Cumulative Sum
cumsum() (cumulative sum): Value of each row is replaced by the sum of all prior rows including this row. 
Char value rows use concatenation as shown below.
'''

import pandas as pd
df = pd.DataFrame({'x': [1, 3, 5],'y': [2, 4, 6],'z':['a','b','k']})
df.cumsum()
```

```python
'''4.1 Practical use: to know the total sum of fee 
No matter how many rows in frame, we can get the last row,,,, and thats the total

Sinceon one var it is a series, we need to convert to a frame
'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
pd.DataFrame(df.Fee.cumsum()).iloc[-1:,:]
```

```python
'''4.2. Cumulative Sum, what would a missing value do to the cumsum, 
the calculation would not make much sense

This is the reason we need to do handle missing values in ML
'''
import pandas as pd
df = pd.DataFrame({'x': [None, 3, 12],'y': [2, 4, 6],'z':['a','b','k']})
df.cumsum()
```

```python
'''4.3 You can compute the sum column wise by changing the axis ...
axis=1 would helpin navigating column wise.

All the values should be of same type.. else would be an issue
'''
import pandas as pd
df = pd.DataFrame({'x': [4, 3, 12],'y': [2, 4, 6],'z':[9,6,7]})
df.cumsum(axis=1)
```

```python
'''
5. query()
Compares two columns and return the rows which holds the condition true

The same frame is replaced 

the returned value is a farme itself
'''
import pandas as pd
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 4, 6],'z':['a','b','k']})
print(df)
a=df.query('x > y')
print(a, type(a))
```

```python
'''5.0.0 This new frame can be obtained'''
import pandas as pd
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 4, 6],'z':['a','b','k']})
new=df.query('x > y')
new
```

```python
'''5.0.1* You can even replce the frame with the results of the query
The frame is now overwritten :)
'''
import pandas as pd
df = pd.DataFrame({'x': [9, 9, 5],'y': [2, 4, 6],'z':['a','b','k']})
df.query('x>y',inplace=True)
df
```

```python
'''5.0.2 How to achieve above with another approach
implace=False is the default as well :)
'''
import pandas as pd
df = pd.DataFrame({'x': [9, 9, 5],'y': [2, 4, 6],'z':['a','b','k']})
df=df.query('x>y',inplace=False)
df
```

```python
'''5.1 in can also be used in query
This would get the matching x and y in a LOOKUP
whereever x matchees y, the x would be output
'''
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 5, 6],'z':['a','b','k']})  
df.query('x in y')
```

```python
'''5.1.1 The logic can be applied on 3 columns as well
It would give x row that matches with y and z
'''
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 5, 6],'z':[1,2,5]})
print(df)
df.query('x in y and x in z')
```

```python
'''5.1.2 Guess the output and hows it is different from "in" 
Corresponding rows would come
'''
import pandas as pd
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 5, 6],'z':['a','b','k']})  
df.query('x==y')
```

```python
'''5.2 not in can also be used with query
Guess the output'''
df = pd.DataFrame({'x': [9, 3, 5],'y': [2, 5, 6],'z':['a','b','k']})  
df.query('x not in y')
```

```python
'''6. Cumulative Prod
Gives the cummulative product, all the columns should have numeric values, as the product would not work for chars

Also avoid Missing values

'''
import pandas as pd
df = pd.DataFrame({'x': [1, 3, 5],'y': [2, 4, 6]})
df.cumprod()
```

```python
'''6.1 Cumprod can handle missing values'''
import pandas as pd
df = pd.DataFrame({'x': [1, 3, 5],'y': [None, 4, 2]})
df.cumprod()
```
