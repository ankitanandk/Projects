```python
'''1. Remove all duplicates , just keep the first occurrence. This is similar to the nodupkey in SAS
we use the drop_duplicates function in Pandas
Subset takes the column name
keep=first means, keep the first instance, this is by default
'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
newdf=df.drop_duplicates(subset='Sex')
newdf
```

```python
'''1.0.1 On a saimple data'''
import pandas as pd
df=pd.DataFrame({'Name':['A','B','B'],'Age':[41,51,61]})
df.drop_duplicates(subset=['Name'],keep='first')
```

```python
'''1.1 We can even give keep=first for instance, this is same as above code'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
newdf=df.drop_duplicates(subset=['Sex'],keep='first')
newdf
```

```python
'''1.2 The keep =last would still remove duplicates; however take the last rows, look at the row index'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
newdf=df.drop_duplicates(subset=['Sex'],keep='last')
newdf
```

```python
'''1.2.1 On a simple data'''
import pandas as pd
df=pd.DataFrame({'Name':['A','B','B'],'Age':[41,51,61]})
df.drop_duplicates(subset=['Name'],keep='last')
```

```python
'''2. Remove all duplicates for Sex. This would not even keep the first occurrence,
if the value are repeated, its gone. so in nutshell only those values are kept 
that have a frequency or occurrence of 1, in this case it would return zero rows

This just keeps the unique values and all duplicated are deleted

UNIQUE VALUES WOULD BE KEPT :)
'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
newdf=df.drop_duplicates(subset=['Sex'],keep=False)
newdf
```

```python
'''2.1 For clarity lets try above code on a simple frame, I hope it is clear now :)
Only unique rows are kept
'''

import pandas as pd
df=pd.DataFrame({'Name':['A','B','B'],'Age':[41,51,61]})
newdf=df.drop_duplicates(subset=['Name'],keep=False)
newdf
```

```python
'''3. Let us remove the duplicates by 2 columns and keep first occurence'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
newdf=df.drop_duplicates(subset=['Sex',"ActLevel"],keep='first' )
newdf
```

```python
'''4. Remove duplicates of the complete row,just like nodup in SAS use drop_duplicates without any argument
It would take all variables by default
'''
Import pandas as pd
raw_data = {'first_name': ['Jason', 'Amit', 'Amit'], 'age': [23, 42, 42]}
df = pd.DataFrame(raw_data)
df.drop_duplicates()
```

```python
'''5. Prove Q5 that drop_duplicates() does not remove a row with even a single value mismatching'''
import pandas as pd
df = pd.DataFrame({'first_name': ['Jason', 'Amit', 'Amit'], 'age': [23, 42, 23]})
df.drop_duplicates()
```

```python
'''6. inplace would overwrite the same frame'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
df.drop_duplicates(subset='Sex',inplace=True)
df.head()
```
