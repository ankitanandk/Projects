```python
'''1. how to get the ages between 40 and 43.. the boundary values are included'''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
df[df.Age.between(40, 43)]
```

```python
'''2. You can even exclude the boundary values '''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
df[df.Age.between(40, 43,inclusive=False)]
```

```python
'''3. You can even use on the Object types... they go by the dictionary '''
import pandas as pd
df=pd.read_csv('/Users/AHAAN/Desktop/Amit_Python/9. data Frames/admit.csv')
df[df.ActLevel.between("G", "I",inclusive=False)]
```
